You may wish to use the `np.polyval(a,x)` method to evaluate the polynomial.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,0,-2,0.5,0])
w = np.arange(-2, 2, 0.1)
f = np.polyval(a,w)
plt.plot(w,f)
```

```python
D = np.gradient(f,w)
g = 4*a[0]*w**3 + 3*a[1]*w**2 + 2*a[2]*w**1 + a[3]
plt.plot(w,D,w,g,w,f)
```
